```python
import numpy as np
import keras.utils as image
import cv2 as cv
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model



```

```python
model=load_model('model_sharp_vgg19.h5')
```

```python


def model_predict(img_path, model):
    img_path=input('Enter the testing image path: ')
    print(img_path)
    img = image.load_img(img_path, target_size=(224, 224))
    
    
    x = image.img_to_array(img)
    x=x/255
    x = np.expand_dims(x, axis=0)
    preds = model.predict(x)
    preds=np.argmax(preds, axis=1)
    
    class_result = ['Anthracnose leaf','Healthy leaf']
    result = class_result[int(preds)]
    print(preds)
    print(result)
    plt.figure(figsize=(6,6))
    
    fig = plt.figure(figsize=(6,6))
    ax = fig.add_subplot(111)
    
    ax.set_title('Predicted : '+result, fontsize=15)
    plt.imshow(img)
    
```

```python
img_path = ''
preds = model_predict(img_path, model)
preds
```
